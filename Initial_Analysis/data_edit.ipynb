```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fullData = pd.read_csv("../final_code/data/weekly_cancer.csv")
#myData = fullData[['Quarter', 'HB', 'CancerType','NumberOfEligibleReferrals31DayStandard','NumberOfEligibleReferralsTreatedWithin31Days']]
HBs=pd.read_csv("../final_code/data/HBs.csv")

fullData.rename(columns={"Cancer type": "CancerType"}, inplace=True)
fullData
```

```python
HB_unique = HBs.drop(HBs.dropna(subset=['HBDateArchived']).index)
HB_unique['HBName'] = HB_unique.HBName.replace({' and ':' & '}, regex=True)
HB_unique
```

```python
HB_code=np.asarray(HB_unique['HB'])
HB_name=np.asarray(HB_unique['HBName'])
dic={}
for i in range(len(HB_code)):
    dic[HB_name[i]] = HB_code[i]

fullData['HB'].replace(dic, inplace=True)
print(np.unique(fullData['HB']))
#dataframe = dataframe.groupby(dataframe.index.names).sum()
```

```python
Areas=np.asarray(fullData['Area name'])
HB_code=np.asarray(HB_unique['HB'])
print(HB_code)
HB_name=np.asarray(HB_unique['HBName'])
print(HB_name)

for i in np.arange(len(HB_name)):
    y=np.where(Areas==HB_name[i])
    Areas[y]=HB_code[i]


fullData['HB'] =Areas.tolist()
print(np.unique(fullData['HB']))
```

```python

Reigons = ['S08000015', 'S08000016', 'S08000017', 'S08000019', 'S08000020', 'S08000022', 'S08000024', 'S08000025', 'S08000026', 'S08000028', 'S08000029', 'S08000030', 'S08000031', 'S08000032']
#Reigons= ['NCA','WOSCAN','SCAN']
  # selecting rows based on condition
# selecting rows based on condition
f=fullData[fullData['Area name'].isin(Reigons)]

#f=fullData
```

```python


keep_col = ['HB','CancerType','Sex','Age Group', 'Week Number','Count 2019']
cancerdata_2019 = f[keep_col]
cancerdata_2019.rename(columns={"Count 2019": "Count"}, inplace=True)
keep_col = ['HB','CancerType','Sex','Age Group', 'Week Number','Count 2020']
cancerdata_2020 = f[keep_col]

cancerdata_2020.rename(columns={"Count 2020": "Count"}, inplace=True)
keep_col = ['HB','CancerType','Sex','Age Group', 'Week Number','Count 2021']
cancerdata_2021 = f[keep_col]

cancerdata_2021.rename(columns={"Count 2021": "Count"}, inplace=True)

```

```python
def weeknumber_2019(weekn):
    quarters=[]
    months=[]
    for wn in weekn:
        if wn in np.arange(6):
            month=1
        elif wn in np.arange(6,10):
            month=2
        elif wn in np.arange(10,14):
            month=3
        elif wn in np.arange(14,20):
            month=4
        elif wn in np.arange(20,24):
            month=5
        elif wn in np.arange(24,27):
            month=6
        elif wn in np.arange(27,32):
            month=7
        elif wn in np.arange(32,36):
            month=8
        elif wn in np.arange(36,41):
            month=9
        elif wn in np.arange(41,45):
            month=10
        elif wn in np.arange(45,49):
            month=11
        elif wn in np.arange(49,153):
            month=12
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
    return np.asarray(quarters),np.asarray(months)

def weeknumber_2020(weekn):
    quarters=[]
    months=[]
    for wn in weekn:
        if wn in np.arange(6):
            month=1
        elif wn in np.arange(6,10):
            month=2
        elif wn in np.arange(10,15):
            month=3
        elif wn in np.arange(15,19):
            month=4
        elif wn in np.arange(19,23):
            month=5
        elif wn in np.arange(23,28):
            month=6
        elif wn in np.arange(28,32):
            month=7
        elif wn in np.arange(32,37):
            month=8
        elif wn in np.arange(37,41):
            month=9
        elif wn in np.arange(41,45):
            month=10
        elif wn in np.arange(45,50):
            month=11
        elif wn in np.arange(49,54):
            month=12
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
    return np.asarray(quarters),np.asarray(months)

def weeknumber_2021(weekn):
    quarters=[]
    months=[]
    for wn in weekn:
        if wn in np.arange(6):
            month=1
        elif wn in np.arange(5,9):
            month=2
        elif wn in np.arange(9,14):
            month=3
        elif wn in np.arange(14,18):
            month=4
        elif wn in np.arange(18,23):
            month=5
        elif wn in np.arange(23,27):
            month=6
        elif wn in np.arange(27,31):
            month=7
        elif wn in np.arange(31,36):
            month=8
        elif wn in np.arange(36,40):
            month=9
        elif wn in np.arange(40,44):
            month=10
        elif wn in np.arange(44,49):
            month=11
        elif wn in np.arange(49,153):
            month=12
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
    return np.asarray(quarters),np.asarray(months)
```

```python
weeknumber_2019(np.asarray(cancerdata_2019['Week Number']))
```

```python

quarter,month=weeknumber_2019(np.asarray(cancerdata_2019['Week Number']))
date=month*100+20190000+np.asarray(cancerdata_2019['Week Number'])
month=month+201900
quarter=quarter+201900
cancerdata_2019['Quarter']=quarter.tolist()
cancerdata_2019['Month'] =month.tolist()
cancerdata_2019['Date'] =date.tolist()


quarter,month=weeknumber_2020(np.asarray(cancerdata_2020['Week Number']))
date=month*100+20200000+np.asarray(cancerdata_2020['Week Number'])
month=month+202000
quarter=quarter+202000
cancerdata_2020['Quarter'] =quarter.tolist()
cancerdata_2020['Month'] =month.tolist()
cancerdata_2020['Date'] =date.tolist()

quarter,month=weeknumber_2021(np.asarray(cancerdata_2021['Week Number']))
date=month*100+20210000+np.asarray(cancerdata_2021['Week Number'])
month=month+202100
quarter=quarter+202100
cancerdata_2021['Quarter'] =quarter.tolist()
cancerdata_2021['Month'] =month.tolist()
cancerdata_2021['Date'] =date.tolist()
```

```python

all_years=pd.concat([cancerdata_2019,cancerdata_2020,cancerdata_2021],ignore_index=True)
```

```python
print(np.unique(all_years['HB']))
all_years = all_years.dropna(axis=0)
all_years
```

```python
all_years.to_csv('../final_code/data/new_cancerdata.csv')
```
