```python
def weeknumber_2019(weekn):
    quarters=[]
    months=[]
    weeks=[]
    for wn in weekn:
        if wn in np.arange(6):
            month=1
            week=wn-1
        elif wn in np.arange(6,10):
            month=2
            week=wn-6
        elif wn in np.arange(10,14):
            month=3
            
            week=wn-10
        elif wn in np.arange(14,20):
            month=4
            
            week=wn-14
        elif wn in np.arange(20,24):
            month=5
            
            week=wn-20
        elif wn in np.arange(24,27):
            month=6
            
            week=wn-24
        elif wn in np.arange(27,32):
            month=7
            
            week=wn-27
        elif wn in np.arange(32,36):
            month=8
            
            week=wn-32
        elif wn in np.arange(36,41):
            month=9
            week=wn-36
        elif wn in np.arange(41,45):
            month=10
            week=wn-41
        elif wn in np.arange(45,49):
            month=11
            week=wn-45
        elif wn in np.arange(49,53):
            month=12
            week=wn-49
        weeks.append(week)
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
    return np.asarray(quarters),np.asarray(months),np.asarray(weeks)+1

def weeknumber_2020(weekn):
    quarters=[]
    months=[]
    weeks=[]
    for wn in weekn:
        if wn in np.arange(6):
            week=wn-1
            month=1
        elif wn in np.arange(6,10):
            month=2
            week=wn-6
        elif wn in np.arange(10,15):
            month=3
            week=wn-10
        elif wn in np.arange(15,19):
            month=4
            week=wn-15
        elif wn in np.arange(19,23):
            month=5
            week=wn-19
        elif wn in np.arange(23,28):
            month=6
            week=wn-23
        elif wn in np.arange(28,32):
            month=7
            week=wn-28
        elif wn in np.arange(32,37):
            month=8
            week=wn-32
        elif wn in np.arange(37,41):
            month=9
            week=wn-37
        elif wn in np.arange(41,45):
            month=10
            week=wn-41
        elif wn in np.arange(45,50):
            month=11
            week=wn-45
        elif wn in np.arange(50,54):
            month=12
            week=wn-50
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
        weeks.append(week)
    return np.asarray(quarters),np.asarray(months),np.asarray(weeks)+1

def weeknumber_2021(weekn):
    quarters=[]
    months=[]
    weeks=[]
    for wn in weekn:
        if wn in np.arange(5):
            week=wn-1
            month=1
        elif wn in np.arange(5,9):
            month=2
            week=wn-5
        elif wn in np.arange(9,14):
            month=3
            week=wn-9
        elif wn in np.arange(14,18):
            month=4
            week=wn-14
        elif wn in np.arange(18,23):
            month=5
            week=wn-18
        elif wn in np.arange(23,27):
            month=6
            week=wn-23
        elif wn in np.arange(27,31):
            month=7
            week=wn-27
        elif wn in np.arange(31,36):
            month=8
            week=wn-31
        elif wn in np.arange(36,40):
            month=9
            week=wn-36
        elif wn in np.arange(40,44):
            month=10
            week=wn-40
        elif wn in np.arange(44,49):
            month=11
            week=wn-44
        elif wn in np.arange(49,54):
            month=12
            week=wn-49
        weeks.append(week)
        months.append(month)
        quarters.append(np.int(np.floor(month/4)+1))
    return np.asarray(quarters),np.asarray(months),np.asarray(weeks)+1
```

```python

quarter,month,week=weeknumber_2019(np.asarray(cancerdata_2019['Week Number']))
date=month*100+20190000+week
month=month+201900
quarter=quarter+201900
cancerdata_2019['Quarter']=quarter.tolist()
cancerdata_2019['Month'] =month.tolist()
cancerdata_2019['Date'] =date.tolist()


quarter,month,week=weeknumber_2020(np.asarray(cancerdata_2020['Week Number']))
date=month*100+20200000+week
month=month+202000
quarter=quarter+202000
cancerdata_2020['Quarter'] =quarter.tolist()
cancerdata_2020['Month'] =month.tolist()
cancerdata_2020['Date'] =date.tolist()

quarter,month,week=weeknumber_2021(np.asarray(cancerdata_2021['Week Number']))
date=month*100+20210000+week
month=month+202100
quarter=quarter+202100
cancerdata_2021['Quarter'] =quarter.tolist()
cancerdata_2021['Month'] =month.tolist()
cancerdata_2021['Date'] =date.tolist()
```

```python
all_years=pd.concat([cancerdata_2019,cancerdata_2020,cancerdata_2021],ignore_index=True)
```

```python
# remove Nans
df=all_years
df = df[df['Count'].notna()]

# remove 'all cancer types' from data
df = df[df['CancerType'] != 'All Cancers']

df = df[df['Sex'] != 'All']
df = df[df['Age Group'] != 'All Ages']


# add a numerical variable 1-1 to date and an index variable before/after pandemic
# and population size
quarters=np.asarray(df['Quarter'])
dates=np.asarray(df['Date'])
months=np.asarray(df['Month'])
HBs=np.asarray(df['HB'])


date=np.zeros(len(dates))
quars=df['Date'].unique()


dates_numerical=np.arange(len(quars))

pop_size=np.zeros(len(dates))

pandemic=np.zeros(len(dates))
pandemic_cat = np.zeros(len(dates))
justpandemic=np.zeros(len(dates))
pandemicRecovery=np.zeros(len(dates))

```

```python




for i in np.arange(len(dates)):
    Q=quarters[i]
    HB=HBs[i]
    pop_size[i]=find_popsize(HB,Q,popsize)
    y=np.where(quars==dates[i])
    date[i]=dates_numerical[y] # change if we don't want the nuemrical variable to be integers (what is best?)


xs=np.where(months>202003) # change to decide in which quarter the pandemic begins
zs = np.where(months > 202102) # change to decide where recovery from pandemic begins.
pandemic_cat[xs]=pandemic_cat[xs]+1

justpandemic[xs]=date[xs]-date[np.min(xs)]

pandemicRecovery[zs]=date[zs]-date[np.min(zs)]   
```

```python

# save data

df['PopSize'] =pop_size.tolist()
df['Date'] =date.tolist()
df['Pandemic_cat'] = pandemic_cat.tolist()
df['JustPandemic'] =justpandemic.tolist()
df['PandemicRecovery'] =pandemicRecovery.tolist()

df
```
